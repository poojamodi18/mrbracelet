```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('gemstone_details (2).csv')
df
```

```python
def purity(string):
  if pd.isnull(string):
        return 'N/A'
  purity_map = {
      '18K' : '18K',
      '18 K' : '18K',
      '14K' : '14K',
      '14 K' : '14K',
      '10K' : '10K',
      '10 K' : '10K',
      '9K' : '9K',
      '9 K' : '9K',
      '925' : '925',
  }

  purities = []
  for key in purity_map:
      if key in string:
          purities.append(purity_map[key])
  return list(set(purities))

df['metal_purity'] = df['Description'].apply(purity)

#append 925 to metal_purity only when style_desc contains 925
#df['metal_purity'] = df['metal_purity'].apply(lambda x: x + ['925'] if '925' in df['style_desc'].str.lower() else x)
#df['metal_purity'] = df.apply(lambda row: row['metal_purity'] + ['925'] if '925' in row['style_desc'] else row['metal_purity'], axis=1)

df
```

```python
def silverMetal(string):
  if pd.isnull(string):
    return 'N/A'

  silver_map = {
      'SILVER' : 'Silver',
      'SILV' : 'Silver',
      'ALLO' : 'Aolly',
      'STER' : 'Sterling Silver',
      'GOSI' : 'Gold-Silver',
      'PLAT' : 'Platinum',
      '9K' : 'Gold',
      '10K' : 'Gold',
      '14K' : 'Gold',
      '18K' : 'Gold',
      '925' : 'Sterling Silver',
  }

  silver = []
  for key in silver_map:
    if key in string:
      silver.append(silver_map[key])
  return list(set(silver))

# Apply the function to the "style_desc" column
df['style_metal'] = df['Type'].apply(silverMetal)

# Apply the function to the "style_desc" column only if 'style_metal' is empty
df.loc[df['style_metal'].apply(lambda x: len(x) == 0), 'style_metal'] = df.loc[
    df['style_metal'].apply(lambda x: len(x) == 0), 'Description'].apply(silverMetal)

# Print the modified DataFrame
df
```

```python
def add_metal_info(row):
    metal_info = f"Metal: {', '.join(row['metal_purity'])} {row['metal_color']}"
    return f"{metal_info} {row['details']}"

# Apply the function to update the 'details' column
df['details'] = df.apply(add_metal_info, axis=1)

df
```

```python
df.to_csv("shopify_details.csv", encoding='utf-8', index=False)
```
