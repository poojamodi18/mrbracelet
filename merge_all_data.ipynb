```python
import pandas as pd
import numpy as np
import re
```

```python
main = pd.read_csv('Stock Report By Item Summary.csv')
main
```

```python
df_comp = pd.read_csv('Stock Report By Style Components.csv')
df_comp
```

```python
df_size = pd.read_csv('Stock Report By Item-Size.csv')
df_size
```

```python
aggregated_size_df = df_size.groupby('Item_num')['Size'].max().reset_index()
aggregated_size_df
```

```python
merged_df = pd.merge(main, aggregated_size_df[['Item_num', 'Size']], left_on='style_cd', right_on='Item_num', how='left')

# Drop the redundant 'Item_num' column
merged_df.drop('Item_num', axis=1, inplace=True)

merged_df
```

```python
total_df = pd.merge(main, df_comp[['Item_style_cd', 'Type', 'metal_color', 'Total Gm', 'Ctw']], left_on='style_cd', right_on='Item_style_cd', how='left')

# Drop the redundant 'Item_num' column
total_df.drop('Item_style_cd', axis=1, inplace=True)

total_df
```

```python
total_df.to_csv("combined_data.csv", encoding='utf-8', index=False)
```
