```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('combined_data.csv')
df
```

```python
# Define a function to extract the value of string
def extract_number(string):
    if pd.isnull(string):  # Check if the value is NaN
        return 'N/A'
    match = re.search(r'W', string)
   # return 'White Gold' if match else None  # Return the matched color or None
    if match:
      return 'White Gold'
    else:
      match = re.search(r'Y', string)
      if match:
        return 'Yellow Gold'
      else:
        match = re.search(r'R', string)
        if match:
          return 'Rose Gold'
        else:
          match = re.search(r'L', string)
          if match:
            return 'Yollow & White Gold'
          else:
            match = re.search(r'H', string)
            if match:
              return 'Rose & White Gold'
            else:
              match = re.search(r'T', string)
              if match:
                return 'Two Tone'
              else:
                match = re.search(r'NA', string)
                if match:
                  return 'N/A'
                else:
                  match = re.search(r'E', string)
                  if match:
                    return 'Three Tone'
                  else:
                    match = re.search(r'3', string)
                    if match:
                      return 'Three Tone'
                    else:
                      match = re.search(r'S', string)
                      if match:
                        return ''
                      else:
                        match = re.search(r'RY', string)
                        if match:
                          return 'Rose & Yellow Gold'
                        else:
                          match = re.search(r'P', string)
                          if match:
                            return ''
                          else:
                            None



# Apply the function to the "style_desc" column
df['metal_color'] = df['metal_color'].apply(extract_number)

# Print the modified DataFrame
df
```

```python
df.to_csv("combined_data_color.csv", encoding='utf-8', index=False)
```
