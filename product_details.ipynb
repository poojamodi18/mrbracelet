```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('combined_data_color.csv')
df
```

```python
def stoneType(string):
  if pd.isnull(string):
    return 'N/A'

  type_map = {
      'RB' : 'Ruby',
      'CZ' : 'Cubic Zirconic',
      'CUBIC' : 'Cubic Zirconic',
      'LGD' : 'Lab-Grown Diamond',
      'TAN' : 'Tanzanite',
      'TANZ' : 'Tanzanite',
      'TPZ' : 'Topaz',
      'SP' : 'Sapphire',
      'SAP' : 'Sapphire',
      'QRTZ' : 'Quartz',
      'CITRIN' : 'Citrine',
      'CTN' : 'Citrine',
      'TOURMALINE' : 'Tourmaline',
      'AMETHYST' : 'Amethyst',
      ' AM' : 'Amethyst',
      'AMETH' : 'Amethyst',
      'EMRD' : 'Emerald',
      'TRQ' : 'Turquoise',
      'TR ' : 'Tourmaline',
      'IOLITE' : 'Iolite',
      'AQ' : 'Aquamarine',
      'PERIDOT' : 'Peridot',
      'PRL': 'Pearl',
      'MRG' : 'Morganite',
      'TURQUOISE' : 'Turquoise',
      'TRQ' : 'Turquoise',
      'GARNET' : 'Garnet',
      'GRNT' : 'Garnet',
      'TREATED' : 'Treated Diamond',
      'JADE' : 'Jade',
      'MG' : 'Morganite',
      'MNST' : 'Moonstone',
      'AGT' : 'Agate',
      'CHNY' : 'Chalcedony',
      'CRY' : 'Crystal',
      'ILT' : 'Iolite',
      'KNZ' : 'Kunzite',
      'LPS' : 'Lapis',
      'MLCT' : 'Malachite',
      'MOP' : 'Mother of Pearl',
      'ONX' : 'Onyx',
      'ONYX' : 'ONYX',
      'OPAL' : 'Opal',
      #'PER' : 'Peridot',
      'PRTE' : 'Prehnite',
      'SNL' : 'Spinel',
      'TRLN' : 'Tourmaline',
      'TSRT' : 'Tsavorite',
  }

  types = []
  for key in type_map:
    if key in string:
      types.append(type_map[key])
  return list(set(types))

# Apply the function to the "style_desc" column
df['gemstone'] = df['Description'].apply(stoneType)

# check if stone_type is empty if it is empty and stone_shape has value then add 'Diamond' in stone_type
condition1 = (df['Description'].str.contains('ST') | df['Description'].str.contains('D')) & (df['gemstone'].apply(lambda x: isinstance(x, list) and len(x) == 0))
df.loc[condition1, 'gemstone'] = 'Diamond'


# Print the modified DataFrame
df
```

```python
def stoneShape(string):
  if pd.isnull(string):
    return 'N/A'

  shape_map = {
      'ASH': 'Asscher',
      'ASC': 'Asscher',
      'ASSCHER' : 'Asscher',
      'CU': 'Cushion',
      'CUSH' : 'Cushion',
      'MQ': 'Marquise',
      'RD': 'Round',
      'PE': 'Pear',
      'PD': 'Pear',
      'TZ': 'Trapezoid',
      'BG': 'Baguette',
      'OV':'Oval',
      'ECUT':'Emerald',
      'EC':'Emerald',
      'EMCD' : 'Emerald',
      'EMCUT' : 'Emerald',
      'SC':'Round Single Cut',
      'PR' : 'Princess',
      'RAD' : 'Radiant',
      'ROSCUT' : 'Rose Cut Pear',
      'ICE' : 'Slice Diamond',
      'TRP' : 'Trapezoid',
      'TRA' : 'Trapezoid',
      'CAD' : 'Cadillac',
      'RS' : 'Rose Cut'

  }

  shapes = []
  for key in shape_map:
    if key in string:
      shapes.append(shape_map[key])
  return list(set(shapes))


# Apply the function to the "style_desc" column
df['shape_setting'] = df['Description'].apply(stoneShape)

#if style_desc contains ST and D and stone shape is empty then add round in stone shape
condition = (df['Description'].str.contains('ST') | df['Description'].str.contains('D')) & (df['shape_setting'].apply(lambda x: isinstance(x, list) and len(x) == 0))

df.loc[condition, 'shape_setting'] = 'Round'


# Print the modified DataFrame
df
```

```python
# Define a function to extract the number before "ST" from the string
def extract_number(string):
    if pd.isnull(string):  # Check if the value is NaN
        return None
    match = re.search(r'(\d+)\s+ST', string)  # Search for a number followed by "ST"
    return match.group(1) if match else None  # Return the matched number or None

# Apply the function to the "style_desc" column
df['stone_no'] = df['Description'].apply(extract_number)

# Print the modified DataFrame
df
```

```python
# Function to create the 'details' string
def create_details(row):
    return f"Gemstone 1: {row['gemstone']} (Quantity: {row['stone_no']}, Shape & Cutting Style: {row['shape_setting']}, Weight: {row['Ctw']} ct.)"

# Apply the function to create the 'details' column
df['details'] = df.apply(create_details, axis=1)

df
```

```python
df.to_csv("gemstone_details.csv", encoding='utf-8', index=False)
```

```python
def create_details(row):
    if isinstance(row['shape_setting'], list):
        details = []
        for idx, shape in enumerate(row['shape_setting'], start=1):
            details.append(f"Gemstone {idx}: {row['gemstone']} (Quantity: {row['stone_no']}, Shape & Cutting Style: {shape}, Weight: {row['Ctw']} ct.)")
        return ' '.join(details)
    else:
        return f"Gemstone 1: {row['gemstone']} (Quantity: {row['stone_no']}, Shape & Cutting Style: {row['shape_setting']}, Weight: {row['Ctw']} ct.)"

# Apply the function to create the 'details' column
df['details'] = df.apply(create_details, axis=1)

df
```
